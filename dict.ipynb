```python
elements=['a','b','c','a','a','b','c','c']
ele_count={}

for item in elements:
    if item in ele_count:
        ele_count[item]+=1 
    else:
        ele_count[item]=1
print(ele_count)
```

```python
elements=['a','b','a', 'a', 'c','c']
d={item:elements.count(item) for item in elements }
print(d)
```
