# Sequences (list, tuple, string)

indexing

```python
x='frog'
print(x[1])
```

```python
x=(12,32,45,78)
print(x[2])
```

Slicing

```python
# slice out some portion using indexing
x='computer' # indexing--> 0 to 7, total=8
#  01234567
```

```python
start=0
end=8
step=2
#print(x[start:end+1:step]) default step=1
print(x[start:end:step])
```

```python
start=4
print(x[start:]) # end is missing, step is missing (default=1)
```

```python
end=8
print(x[:end])
```

```python
# -ve indexing
x='computer'
#      0   1   2   3   4   5   6   7 (+ve indexing)
#      c   o   m   p   u   t   e  r
#     -8  -7  -6  -5  -4  -3  -2  -1 (-ve indexing)
print(x[-5])
print(x[3])
```

```python
print(x[-3:-1])  # start=-3, end=-1
```

```python
print(x[:-2])
```

```python
print(x[2:-4])
```

```python
print(x[2:-4:2])
```

iterating

```python
x=[10,20,30,40,50]
for item in x:
    print(item) # iterating
```

```python
x=[4,1,7,3,9,5,6]
for index,item in enumerate(x):
    print(index,item)
```
