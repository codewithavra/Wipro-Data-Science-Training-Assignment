# Anonymous function

```python
def func(a):
    return a+10

func(2)
```

```python
f=lambda a:a+10  # arg:return value
f(2)
```

```python
y=lambda x:x**2
y(3)
```

```python
y=lambda x,z:z+x**2
y(2,3)
```

```python
def myFunc(n):
    return lambda a:a*n
```

```python
d=myFunc(2) # n=2, d=lambda a:a*2
d(3)  # a=3
```

```python
#f(x)=x^2+2x+1
#find f(x) for x=4

def func(x):
    return x**2+2*x+1
func(4)

f=lambda x:x**2+2*x+1
f(4)
```
