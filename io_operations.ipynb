```python
f=open("data.txt","w")
```

```python
# open
f.write("Python") # old contents are deleted
f.close()
```

```python
# Append
f=open("data.txt","a")
f.write("\nJava")
f.close()
```

```python
# cretae a file
f=open("data1.txt","x") # to create new file
```

```python
# To read the entire file
f=open("data.txt","r")
print(f.read())
f.close()
```

```python
# read specific number of characters
f=open("data.txt","r")
print(f.read(6))
f.close()
```

```python
# Read one line
f=open("data.txt","r")
print(f.readline())
print(f.readline())
f.close()
```

```python
# Read list of lines
f=open("data.txt","r")
print(f.readlines())
f.close()
```

```python
# read using loop
f=open("data.txt","r")

for line in f:
    print(line)
f.close()
```

### Using with statement

```python
# manual close() is not required here
with open("data.txt","r") as f:
    print(f.read())
```

### read and write both

```python
with open("sample.txt","w") as f:
    f.write("Python\n")
    f.write("JAVA\n")
    f.write("C++")

with open("sample.txt","r") as f:
    print(f.read())
```

```python
try:
    f=open("test.txt","r")
    print(f.read())
    f.close()
except FileNotFoundError:
    print("File not found")
```

# Delete files

```python
import os
os.remove("sample.txt")
```

### check before delete

```python
filename="sample.txt"
if os.path.exists(filename):
    os.remove(filename)
else:
    print("File does not exist")
```

### Using exception

```python
try:
    os.remove("sample.txt")
except FileNotFoundError:
    print("File does not exist")
except PermissionError:
    print("Permission denied")
```

### Delete multiple files

```python
f=open("a.txt","w")
f=open("b.txt","w")
f=open("c.txt","w")
```

```python
import os
files=["a.txt","b.txt","c.txt"]

for file in files:
    try:
        os.remove(file)
        print(file," deleted")
    except FileNotFoundError:
        print("File does not exist")
```
