# Keyword arguments

```python
def printinfo(name,roll):
    print("Name:",name)
    print("Roll:",roll)
    return
```

```python
printinfo("kaushik",45)
```

```python
printinfo(30,"kaushik")
```

```python
printinfo(roll=30,name="kaushik")
```

```python
printinfo(roll=30,"kaushik")
```

```python
printinfo(roll=30)
```

# Default arguments

```python
def printinfo(name,roll=20):
    print("Name:",name)
    print("Roll:",roll)
    return
```

```python
printinfo(name="kaushik")
```

```python
printinfo(name="kaushik",roll=40)
```

# Variable length arguments

```python
def printinfo(arg1,arg2,*vararg):
    print(arg1)
    print(arg2)
    #print(vararg)
    for var in vararg:
        print(var)
    return
```

```python
printinfo(10,20)
```

```python
printinfo(10,20,30,40,50,1,1,2,3,4,5,6,7,8)
```
