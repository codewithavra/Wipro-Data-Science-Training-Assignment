# Loops

```python
c=0
while c<5:
    c=c+1
    print('Python')
```

```python
c=0
while c<5:
    c=c+1
    print('Inside the loop')
else:
    print('Outside the loop')
```

```python
n=10
for i in range(3,n,2): # range(start,end,step)
    print(i)
```

Iterate over list, tuple, string dictionary

```python
s='python'
for ch in s:
    print(ch)
```

```python
l=['a','b','c']
for item in l:
    print(item)
```

```python
l=['a','b','c']
for item in l:
    print(item)
else:
    print('I am in ELSE')
```

```python
#nested loop
for i in range(1,5):
    for j in range(i):
        print(i,end=' ')
    print()
```

```python
str='abcdefghijkl'
# print all elements except d and i
for letter in str:
    if letter=='d' or letter=='i':
        continue
    print(letter)
```

```python
str='abcdefghijkl'

for letter in str:
    if letter=='d' or letter=='i':
        break
    print(letter)
```

```python
str='abcdefghijkl'

for letter in str:
    pass # do nothing, it introduces delay
print(letter)
```
