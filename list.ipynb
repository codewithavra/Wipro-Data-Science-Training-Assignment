# List Comprehension

```python
x=[]
for i in range(5):
    if i>2:
        a=i**2
        x.append(a)
print(x)
```

```python
x=[i**2 for i in range(5) if i>2] # Comprehension
print(x)
```

```python
x=[i if i%2==0 else i for i in range(10) if i>4] # Comprehension
print(x)
```
