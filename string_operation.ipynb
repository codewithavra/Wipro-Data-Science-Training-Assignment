# Concatenation and repetition

```python
a="Hello"
b="world"
print(a+' '+b)
print(a*5)
```

```python
# membership
s="Python"
print("Py" in s)
print("J" in s)
```

# Case conversion

```python
s="PyThOn"
```

```python
print(s.lower())
print(s.upper())
s="i love python"
print(s.title())
print(s.capitalize())
print(s.swapcase())
```

```python
# Whitespace
s="    Hello world    "
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
# Serach
s="abcdefee"
print(s.find('z'))
print(s.count('e'))
print(s.index('z'))
```

```python
s="adsfsgdjdkfjriuryry"
ch='z'

if s.find(ch)>-1:
    print(s.index(ch))
else:
    print(ch+" not present")
```

```python
# Replace

s="I like Python"
print(s.replace('Python','Java'))
```

```python
# split and join
s="apple,banana,mango"
lst=s.split(",")
print(lst)
print("-".join(lst))
```

```python
s="apple,banana,mango"
print(s.split(",")[1])  # banana
print(s.split(",")[1][::-1])
#---------------------   ---------------
#      banana             reverse the string
```

```python
# starts/end
s="Pyrhon programming"
print(s.startswith("Py"))
print(s.endswith("ing"))
```

```python
# Testing methods
s1="python"
s2="12345"
s3="Python3"
print(s1.isalpha())
print(s2.isdigit())
print(s1.islower())
print(s3.isupper())
print(s1.upper().isupper())
print(" ".isspace())
```

```python
# Formatting
name="Alice"
age=25
print(f"{name} is {age} years old")
print("{} is  {}".format(name,age))
print("%s is %d years old"%(name,age))
```

```python
# Alignment
s="Python"
print(s.center(20,'-'))
print(s.ljust(20,'-'))
print(s.rjust(20,'-'))
```

```python
# Escape characters
print("Hello\nworld")
print("Hello\tworld")
```
