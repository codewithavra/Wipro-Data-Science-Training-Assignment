```python
set1={1,2,3,4,5}
set2={1,2,3}
```

```python
# intersection (AND)
set1 & set2
```

```python
# union (OR)
set1 | set2
```

```python
set1={1,2,3,4}
set2={3,4,5,6}
```

```python
# symmetric difference (XOR)
# returns the elements that are in either set, but not in both
set1 ^ set2
```

```python
# difference  
print(set1-set2) #in set1 but not in set2
print(set2-set1) #in set2 but not in set1
```

```python
# subset (set2 contains set1)
set1={1,2}
set2={1,2,3,4}
set1<=set2
```

```python
# superset (set1 contains set2)
set1>=set2
```
