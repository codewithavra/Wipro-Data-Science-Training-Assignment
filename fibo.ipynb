```python
fibo_cache={}

def fibo(n):
    if n in fibo_cache:
        return fibo_cache[n]
    if n==1:
        return 1
    elif n==2:
        return 1
    elif n>2:
        value = fibo(n-1)+fibo(n-2)

    fibo_cache[n]=value
    return value

for n in range(1,11): #n=1,2,3,...,9
    print(n,':',fibo(n))
```
